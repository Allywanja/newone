```python
import gradio as gr
import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.huggingface_api import HuggingFaceInferenceAPI
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

# 1. Setup Models using the Secret Token
# Access the HF_TOKEN you will set in your Space Settings
hf_token = os.getenv("HF_TOKEN")

Settings.llm = HuggingFaceInferenceAPI(
    model_name="mistralai/Mistral-7B-Instruct-v0.3",
    token=hf_token
)
Settings.embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")

# 2. Load the PDF and build the Index
# Ensure "transformers.pdf" is uploaded to your Space files
documents = SimpleDirectoryReader(input_files=["transformers.pdf"]).load_data()
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# 3. Define the Chat Logic
def predict(message, history):
    response = query_engine.query(message)
    return str(response)

# 4. Launch the Gradio Interface
demo = gr.ChatInterface(
    fn=predict,
    title="Attention Is All You Need: Research Bot",
    description="Ask questions about the original Transformer architecture paper."
)

if __name__ == "__main__":
    demo.launch()
```
